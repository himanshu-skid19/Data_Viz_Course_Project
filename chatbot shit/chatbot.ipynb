```python
def import_documents_from_file(input_file: str = "attraction_documents.jsonl") -> List[Document]:
    """
    Import LlamaIndex documents from a text file in JSONL format
    
    Args:
        input_file: Path to the input file
        
    Returns:
        List of LlamaIndex documents
    """
    input_path = Path(input_file)
    
    # Check if file exists
    if not input_path.exists():
        print(f"Error: File {input_path} does not exist")
        return []
    
    documents = []
    
    # Read documents from file
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            try:
                # Parse JSON line
                doc_dict = json.loads(line.strip())
                
                # Create LlamaIndex document
                doc = Document(
                    text=doc_dict['text'],
                    metadata=doc_dict.get('metadata', {}),
                    id_=doc_dict.get('id')
                )
                
                documents.append(doc)
            except json.JSONDecodeError:
                print(f"Error: Could not decode JSON line: {line}")
                continue
    
    print(f"Imported {len(documents)} documents from {input_path}")
    return documents
```

```python

def create_documents_from_tourism_text(text_content: str) -> List[Document]:
    """
    Create Document objects from tourism text content.
    
    Args:
        text_content: String containing the tourism text data
        
    Returns:
        List of Document objects
    """
    # Split the text into sections based on blank lines between paragraphs
    sections = re.split(r'\n\n+', text_content)
    
    # Group sections by attractions - we'll consider sections that are separated 
    # by multiple blank lines to be different attractions
    attractions = []
    current_attraction = []
    
    for section in sections:
        if section.strip():  # Skip empty sections
            current_attraction.append(section.strip())
        elif current_attraction:  # If we have content and encounter empty section, complete the attraction
            attractions.append("\n\n".join(current_attraction))
            current_attraction = []
    
    # Don't forget the last attraction if there's content
    if current_attraction:
        attractions.append("\n\n".join(current_attraction))
    
    # Create documents for each attraction
    documents = []
    
    # Define categories for attractions
    categories = [
        "historical_sites", 
        "cultural_tours", 
        "recommended_experiences", 
        "spiritual_destinations",
        "architectural_marvels",
        "natural_wonders",
        "food_experiences"
    ]
    
    # Define ratings for attractions - higher for more famous places
    ratings_map = {
        "Taj Mahal": 4.9,
        "Delhi": 4.5,
        "Amritsar": 4.7,
        "Golden Temple": 4.9,
        "Manali": 4.6
    }
    
    # Map attractions to cities and locations
    city_map = {
        "Taj Mahal": {"city": "agra", "state": "uttar_pradesh", "country": "india"},
        "Delhi": {"city": "delhi", "state": "delhi", "country": "india"},
        "Amritsar": {"city": "amritsar", "state": "punjab", "country": "india"},
        "Golden Temple": {"city": "amritsar", "state": "punjab", "country": "india"},
        "Manali": {"city": "manali", "state": "himachal_pradesh", "country": "india"}
    }
    
    for idx, attraction_text in enumerate(attractions):
        # Create a unique ID
        doc_id = str(uuid.uuid4())
        
        # Extract the first line or first few words as a title
        first_line = attraction_text.split('\n')[0].strip()
        title = first_line[:50]  # Use first 50 chars as title
        
        # Determine attraction name and location
        attraction_name = None
        location = {"city": "unknown", "state": "unknown", "country": "india"}
        
        # Try to identify the attraction
        for key in city_map:
            if key.lower() in attraction_text.lower()[:500]:  # Check in first 500 chars
                attraction_name = key
                location = city_map[key]
                break
        
        if not attraction_name:
            # Fallback to the first capitalized phrase
            match = re.search(r'([A-Z][a-z]+(?: [A-Z][a-z]+)*)', attraction_text)
            if match:
                attraction_name = match.group(0)
        
        # If still no name, use generic
        if not attraction_name:
            attraction_name = f"Indian Attraction {idx+1}"
        
        # Determine category based on content
        category = random.choice(categories)
        if "temple" in attraction_text.lower():
            category = "spiritual_destinations"
        elif "food" in attraction_text.lower() or "cuisine" in attraction_text.lower():
            category = "food_experiences"
        elif "heritage" in attraction_text.lower() or "historic" in attraction_text.lower():
            category = "historical_sites"
        
        # Determine rating
        rating = ratings_map.get(attraction_name, round(random.uniform(4.0, 4.9), 1))
        

        

        # Create metadata
        metadata = {
            'doc_type': 'attraction',
            'category': category,
            'name': attraction_name,
            'attraction_id': int(random.randint(1000, 9999)),
            'country': location['country'],
            'state': location['state'],
            'city': location['city'],

            'unesco_site': "taj mahal" in attraction_name.lower(),
            'best_time_to_visit': get_best_time_to_visit(attraction_text)
        }
        
        # Create text content in the desired format
        formatted_text = f"""Type: Tourist Attraction
Name: {attraction_name}
Location: {location['city'].capitalize()}, {location['state'].replace('_', ' ').capitalize()}, {location['country'].capitalize()}
Category: {category.replace('_', ' ').title()}
Best Time to Visit: {metadata['best_time_to_visit']}

Description:
{attraction_text}
"""
        
        # Create Document object
        doc = Document(
            id_=doc_id,
            metadata=metadata,
            text=formatted_text
        )
        
        documents.append(doc)
    
    return documents


def get_best_time_to_visit(text):
    """Extract or determine best time to visit from text"""
    seasons = ["Spring", "Summer", "Autumn", "Winter"]
    months = ["January", "February", "March", "April", "May", "June", 
              "July", "August", "September", "October", "November", "December"]
    
    # Try to find mentions of best time
    for season in seasons:
        if f"best time" in text.lower() and season.lower() in text.lower():
            return season
    
    for month in months:
        if f"best time" in text.lower() and month.lower() in text.lower():
            return month
    
    # If not found, make an educated guess based on location
    if "manali" in text.lower():
        return "April to June"
    elif "taj mahal" in text.lower():
        return "October to March"
    elif "delhi" in text.lower():
        return "October to March"
    elif "amritsar" in text.lower() or "golden temple" in text.lower():
        return "November to March"
    
    # Default
    return "October to March"
```

```python
def get_max_document_length(documents: List[Document]) -> dict:
    """
    Find the maximum text length among a list of LlamaIndex documents,
    and return detailed information about the longest document.
    
    Args:
        documents: List of LlamaIndex Document objects
        
    Returns:
        Dictionary with information about max length and the document with maximum length
    """
    if not documents:
        return {"max_length": 0, "max_doc_id": None, "max_doc_name": None}
    
    max_length = 0
    max_doc = None
    
    for doc in documents:
        # Get the text from the document's text_resource
        doc_text = doc.text_resource.text if doc.text_resource else ""
        text_length = len(doc_text)
        
        if text_length > max_length:
            max_length = text_length
            max_doc = doc
    
    # Return information about the maximum length document
    result = {
        "max_length": max_length,
        "max_doc_id": max_doc.id_ if max_doc else None,
        "max_doc_name": max_doc.metadata.get("name", "Unknown") if max_doc else None,
        "character_count": max_length,
        "word_count": len(max_doc.text_resource.text.split()) if max_doc and max_doc.text_resource else 0,
        "line_count": len(max_doc.text_resource.text.splitlines()) if max_doc and max_doc.text_resource else 0
    }
    
    return result

def get_document_length_stats(documents: List[Document]) -> dict:
    """
    Calculate various statistics about document lengths in a list of LlamaIndex documents.
    
    Args:
        documents: List of LlamaIndex Document objects
        
    Returns:
        Dictionary with length statistics
    """
    if not documents:
        return {
            "count": 0,
            "min_length": 0,
            "max_length": 0,
            "avg_length": 0,
            "total_length": 0
        }
    
    lengths = []
    
    for doc in documents:
        doc_text = doc.text_resource.text if doc.text_resource else ""
        lengths.append(len(doc_text))
    
    return {
        "count": len(documents),
        "min_length": min(lengths),
        "max_length": max(lengths),
        "avg_length": sum(lengths) / len(lengths),
        "total_length": sum(lengths),
        "median_length": sorted(lengths)[len(lengths) // 2]
    }
```

```python
from llama_index.core.query_engine import RetrieverQueryEngine

def query_vector_store(prompt, top_k=2):
    """
    Query the vector store through the query engine to find similar resumes.

    Args:
    - prompt (str): The input prompt to query for similar resumes.
    - top_k (int): The number of similar resumes to retrieve.

    Returns:
    - list[str]: A list of similar resume texts.
    """
    # Use the query engine to query the index with your prompt
    nodes = bm25_retriever.retrieve(
        query_str
    )
    retrieved_nodes = "\n".join(str(node) for node in nodes)
    formatted_prompt = SYSTEM_PROMPT.format(USER_QUERY=query_str, RETRIEVED_NODES=retrieved_nodes_)
    response = model.complete(formatted_prompt)
    

    return response
```
